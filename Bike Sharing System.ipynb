```python
#data modelling
bike1= bike.drop(['instant','dteday','casual','registered'], axis=1)
bike1.head()
```

```python
#Mapping Season to categorical data
bike1.season.replace({1: 'spring',2:'summer',3:'fall',4:'winter'}, inplace = True)
bike1.tail()
```

```python
# mapping weekday to categorical data

bike1.weekday = bike1.weekday.map({0:"Sunday",1:"Monday",2:"Tuesday",3:"Wednesday",4:"Thrusday",5:"Friday",6:"Saturday"})
```

```python
# mapping weathersit to categorical data

bike1.weathersit = bike1.weathersit.replace({1:'Clear',2:'Mist & Cloudy',3:'Light Snow & Rain',4:'Heavy Snow & Rain'})
bike1.head()
```

```python
# mapping mnth with month categorical data

import calendar
bike1['mnth'] = bike1['mnth'].apply(lambda x: calendar.month_abbr[x])
bike1.head()
```

```python
from sklearn.metrics import r2_score
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
#create dummy variable
bike_3 = bike1[['season','mnth','weathersit','weekday']]
bike_3 = pd.get_dummies(bike_3,drop_first=True)
# Adding dummies back to the orginal data frame
bike1 = pd.concat([bike_3,bike1], axis = 1)
bike1.head()
```

```python
# Dropping columns for which dummy variables were created

bike1.drop(['season','mnth','weathersit','weekday'], axis = 1, inplace = True)
```

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
#scaling again
scaler = MinMaxScaler()
bike1.columns
```

```python
# Train dataframe to have 70%, Test dataframe to have 30% with random values = 100%

np.random.seed(0)
bike_train, bike_test  = train_test_split(bike1, train_size=0.7, test_size=0.3, random_state= 100)
```

```python
# Checking the shape of Train and Test dataframe
print(bike_train.shape, bike_test.shape)
```

```python
#checking training dataset
plt.figure(figsize = (16, 15))
sns.heatmap(bike_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

```python
sns.regplot( x= 'temp', y= 'cnt', data=bike1)
plt.show()
```

```python
#separate x and y
y_train = bike_train.pop('cnt')
X_train = bike_train
```

```python
# Checking the shape
print(y_train.shape, X_train.shape)
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Adding a constant 
X_train_lm = sm.add_constant(X_train[['temp']])

#Create a fit model first
lr1 = sm.OLS(y_train, X_train_lm).fit()

# Checking the parameters
lr1.params
```

```python
# summary of the linear regression model
lr1.summary()
```

With Temp R s1quare is 41%,so consider more variables also

```python
lm = LinearRegression()
lm.fit(X_train, y_train)

# Checking the top 15 variables
rfe = RFE(lm, 15)
rfe = rfe.fit(X_train, y_train)
```

```python
col = X_train.columns[rfe.support_]
```

```python
X_train.columns[~rfe.support_]
```

```python
X_train = X_train[col]
```

```python
# Adding constant
X_train_lm= sm.add_constant(X_train)

lm1 = sm.OLS(y_train,X_train_lm).fit()

lm1.summary()
```

```python
# Checking VIF values
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values,i) for i in range(X_train.shape[1])] 
vif['VIF'] = round(vif['VIF'], 2)
vif.sort_values(by= 'VIF', ascending = False, inplace= True)
vif
```

```python
y_train_cnt = lm1.predict(X_train_lm)
```

```python
# Plot the histogram of the error terms

plt.figure(figsize = (10, 6))
sns.distplot((y_train - y_train_cnt), bins = 40)
plt.show()
```

```python
y_test= bike_test.pop('cnt')
X_test = bike_test
X_test.head()
```

```python
# Selecting variables that were part of final model

col_test= X_train.columns
X_test = X_test[col_test]
X_test.columns

# Adding constant
X_test_lm = sm.add_constant(X_test)

# Making predictions using the final model (lm2)

y_pred = lm1.predict(X_test_lm)
y_pred
```

```python
# Plotting y_test and y_pred to understand the spread.
fig = plt.figure()
plt.scatter(y_test,y_pred, alpha= 0.5)
#plt.plot(y_test,y_pred )

fig.suptitle('y_test vs y_pred', fontsize=15)              # Plot heading 
plt.xlabel('y_test', fontsize=12)                          # X-label
plt.ylabel('y_pred', fontsize=12)                          # Y-label
plt.show()
```

```python
#Rsquare value
r2= r2_score(y_test, y_pred)
r2
```

```python
# n is number of rows in X

n = X_test.shape[0]


# Number of features (predictors, p) is the shape along axis 1
p = X_test.shape[1]

# We find the Adjusted R-squared using the formula

adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
adjusted_r2
```
